```python
# Librerias
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
# Leer datos con pandas
ds = pd.read_csv('dataset_argentina.csv', delimiter=';')
ds
```

```python
colores_barras = ['C0','C1','C2','C3']
colores_fondos = ['azure', 'mistyrose', 'honeydew']
categorias = ds['Categoria']
tipos = ds.drop(columns=['Categoria'])
lista_tipos = tipos.columns.tolist()
for index, tipo in enumerate(lista_tipos):
    fig, ax = plt.subplots()
    barh = ax.barh(categorias, tipos[tipo], color=colores_barras)
    ax.set_title(f'{tipo} | {sum(tipos[tipo])} en total', fontsize='xx-large', fontweight='bold')
    fig.set_facecolor(colores_fondos[index])
    ax.set_facecolor(colores_fondos[index])
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    plt.yticks(fontsize='large')
    plt.xticks(fontsize='large')
    ax.xaxis.set_major_formatter(ticker.EngFormatter())
    list_label_bar=[]
    tipo_valores = tipos[tipo]
    for x in tipo_valores:
        label_bar = f'{str(round(x/sum(tipo_valores)*100,2))}% ({x})'
        list_label_bar.append(label_bar)
    ax.bar_label(barh, list_label_bar, fmt='%d', fontsize='x-large')
```
